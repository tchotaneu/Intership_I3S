```python
import numpy as np
# Exemple d'utilisation :
data = np.loadtxt("sauvegarde/fichierZip/mane_zip3/embedding/Embedding_Entrainer_epoques_14_.txt")
vector_names = data[:, 0].astype(int)
```

```python
#['euclidean', 'l2', 'l1', 'manhattan', 'cityblock', 'braycurtis', 'canberra', 'chebyshev', 'correlation', 'cosine', 'dice', 'hamming', 'jaccard', 'kulsinski', 'mahalanobis', 'matching', 'minkowski', 'rogerstanimoto', 'russellrao', 'seuclidean', 'sokalmichener', 'sokalsneath', 'sqeuclidean', 'yule', 'wminkowski', 'nan_euclidean', 'haversine']
from sklearn.manifold import TSNE

def reduce_with_tsne(data, n_components=2, perplexity=30, learning_rate=100, n_iter=1000, random_state=None, metric='cosine', method='exact'):
    """
    Réduit la dimension des données en utilisant t-SNE.

    :param data: Les données à réduire en dimension (un tableau numpy).
    :param n_components: Le nombre de dimensions dans lesquelles réduire les données (par défaut, 2).
    :param perplexity: La perplexité (par défaut, 30).
    :param learning_rate: Le taux d'apprentissage (par défaut, 200).
    :param n_iter: Le nombre d'itérations (par défaut, 1000).
    :param random_state: Graine aléatoire pour la reproductibilité (par défaut, None).
    :param metric: Métrique de distance (par défaut, 'euclidean').
    :param method: Méthode d'optimisation ('barnes_hut' ou 'exact', par défaut 'barnes_hut').

    :return: Les données réduites en dimension.
    """
    tsne = TSNE(n_components=n_components, perplexity=perplexity, learning_rate=learning_rate, n_iter=n_iter, random_state=random_state, metric=metric, method=method)
    reduced_data = tsne.fit_transform(data)
    return reduced_data
```

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  

def plot_tsne_2D(reduced_data, labels=None, title="t-SNE Visualization"):
    """
    Visualise les données réduites en dimension à l'aide de t-SNE.

    :param reduced_data: Les données réduites en dimension.
    :param labels: Étiquettes (le cas échéant) pour colorer les points.
    :param title: Titre du graphique.
    """
    plt.figure(figsize=(8, 6))

    if labels is None:
        plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
    else:
        unique_labels = set(labels)
        for label in unique_labels:
            plt.scatter(reduced_data[labels == label, 0], reduced_data[labels == label, 1], label=label)
        plt.legend()

    plt.title(title)
    plt.xlabel("Dimension 1")
    plt.ylabel("Dimension 2")
    plt.show()

# Utilisation de la fonction pour visualiser les données réduites en dimension



def plot_tsne_3D(reduced_data, labels=None, title="t-SNE 3D Visualization"):
    """
    Visualise les données réduites en trois dimensions à l'aide de t-SNE.

    :param reduced_data: Les données réduites en dimension (3D).
    :param labels: Étiquettes (le cas échéant) pour colorer les points.
    :param title: Titre du graphique.
    """
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')  # Créer une projection 3D

    if labels is None:
        ax.scatter(reduced_data[:, 0], reduced_data[:, 1], reduced_data[:, 2])
    else:
        unique_labels = set(labels)
        for label in unique_labels:
            ax.scatter(reduced_data[labels == label, 0], reduced_data[labels == label, 1], reduced_data[labels == label, 2], label=label)
        ax.legend()

    ax.set_title(title)
    ax.set_xlabel("Dimension 1")
    ax.set_ylabel("Dimension 2")
    ax.set_zlabel("Dimension 3")
    plt.show()
```

```python

from sklearn.cluster import OPTICS
from collections import defaultdict

def cluster_with_labels(data, min_samples=3, xi=0.5):
    # Sélectionnez les caractéristiques (ignorez la colonne des labels/identifiants)
    X = data[:, 1:]

    # Créez une instance de l'algorithme OPTICS
    optics = OPTICS(min_samples=min_samples, xi=xi, metric='cosine')

    # Appliquez OPTICS sur vos données
    optics.fit(X)

    # Extrait les labels de clusters
    labels = optics.labels_

    # Créez un dictionnaire pour stocker les labels de la première colonne par cluster
    cluster_labels = defaultdict(list)
    for i, label in enumerate(labels):
        if label != -1:  # Ignorez les points qui ne sont pas attribués à un cluster
            cluster_labels[label].append(int(data[i, 0]))

    return dict(cluster_labels)

```

```python

#{129, 37, 680, 809, 330, 873, 143, 120, 30, 895}
import random
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def visualize_clusters(data, cluster_labels):
    # Réduire la dimension des données en utilisant PCA (2 composantes pour une visualisation 2D)
    pca = PCA(n_components=2)
    reduced_data = pca.fit_transform(data[:, 1:])  # Ignorez la première colonne (labels)

    # Créez un dictionnaire pour stocker les couleurs des clusters
    unique_labels = set(cluster_labels)
    cluster_colors = {label: "#" + "%06x" % random.randint(0, 0xFFFFFF) for label in unique_labels}

    # Créez un graphique pour visualiser les clusters
    plt.figure(figsize=(10, 6))
    for i, label in enumerate(cluster_labels):
        if label != -1:  # Ignorez les points non attribués à un cluster
            plt.scatter(reduced_data[i, 0], reduced_data[i, 1] ,c=cluster_colors[label], s=50)

    plt.title("Visualisation des clusters")
    plt.xlabel("Composante principale 1")
    plt.ylabel("Composante principale 2")
    plt.show()
```
